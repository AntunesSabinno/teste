```python
import pandas as pd
```

```python
capitais = [["Acre","AC","800000","Rio Branco",28.6],
["São Paulo","SP","44000000","São Paulo",8.4],
["Rio Grande do Sul","RS","10880506","Porto Alegre",19.8],
["Rio de Janeiro","RJ","16054524","Rio de Janeiro",27.9],
["Paraná","PR","11443208","Curitiba",22.7],
["Bahia","BA","14136417","Salvador",47.1],
["Pernambuco","PE","9058155","Recife",37.8],
["Alagoas","AL","3127511","Maceió",37.9],
["Espírito Santo","ES","3833486","Vitória",29.3]]

df = pd.DataFrame(capitais,columns= ["Estado","UF","População","Capital","Homicídios por 100 mil/hab por estado 2022"])
```

```python
print(df)
```

```python
df["População"]= df.População.astype("int64")    # conversão de tipo de dado, de string para inteiro
```

```python
df.info()
```

```python
df.sort_values(by="Homicídios por 100 mil/hab por estado 2022", ascending = True)    ##ordenação por taxa de Homicídios em ordem crescente
```

```python
df.sort_values(by="Homicídios por 100 mil/hab por estado 2022", ascending = False)   # ordem decrescente
```
